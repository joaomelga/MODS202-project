```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('textfiles/wage1.raw', delim_whitespace=True, header=None)

wage = df[0]
```

```python
plt.hist(wage, 'auto')
```

```python

wage.hist(bins=20)
```

```python
print('mean: ', np.mean(wage))
print('std: ', np.std(wage))
print('max: ', np.max(wage))
print('min: ', np.min(wage))
```

```python
educ = df[1]

print('Covariance: ', np.cov(wage, educ))
print('Correlation: ', np.corrcoef(wage, educ))
```

```python
plt.scatter(wage, educ)
```

```python
isWomen = df[5]
np.sum(isWomen)

womenMeanWage = np.mean(wage[isWomen == 1])
manMeanWage = np.mean(wage[isWomen == 0])

print('Mean W', womenMeanWage)
print('Mean M', manMeanWage)
```

```python
isWomen = df[5]
medianWage = np.median(wage)
womenHigherThanMedian = wage[isWomen == 1][wage > medianWage]

print('Mean W higher', np.mean(womenHigherThanMedian))
```

```python
print(np.percentile(wage, 5))
```

```python
tenures = df[3]
uniqueTenures = np.unique(tenures)
meanWageByTenure = []

for tenure in uniqueTenures:
    meanWageByTenure.append(np.mean(wage[tenures == tenure]))

print(uniqueTenures)
plt.plot(uniqueTenures, meanWageByTenure)
```
